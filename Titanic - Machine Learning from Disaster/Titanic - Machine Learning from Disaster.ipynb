```python
import pandas as pd

data=pd.read_csv("/content/train.csv")
data
```

```python
test=pd.read_csv("/content/test.csv")
test
```

```python
test_ids = test["PassengerId"]
def cleaning(data):
  data=data.drop(["Ticket","Cabin","Name","PassengerId"],axis=1) #axis=1 for column
  # number of siblings & spouses  for SibSp
  # number of parent or children for Parch

  #those columns which have null value and have numerical value
  cols=["SibSp","Parch","Fare","Age"]
  for col in cols:
    data[col].fillna(data[col].median(),inplace=True)#inplace is for overriding existing dataframne

  #for column Embarked if there is null value then we will fill it with "U" means unknown
  data.Embarked.fillna("U", inplace=True)
  return data

```

```python
#driver code for cleaning
data = cleaning(data)
test = cleaning(test)
```
